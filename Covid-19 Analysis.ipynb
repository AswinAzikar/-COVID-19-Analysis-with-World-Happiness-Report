```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sb 
import pandas as pd
```

```python
covid_conf = pd.read_csv(  '/kaggle/input/covid-19-analysis-data/covid19_Confirmed_dataset.csv' )
```

```python
covid_conf.head()
```

```python
covid_conf.shape
```

```python
covid_conf.drop(["Lat","Long","Province/State" ],axis=1,inplace=True)

#
# This Python code is used to remove the "Lat" and "Long" columns from a Pandas DataFrame called "covid_conf":

# covid_conf.drop(...): This part calls the drop() method on the DataFrame covid_conf. The drop() method is used to remove rows or columns from a DataFrame.
# ["Lat", "Long"]: This is a list containing the names of the columns you want to remove. In this case, you're removing the columns labeled "Lat" and "Long".
# axis=1: This argument specifies that you want to drop columns. If you set axis=0, you would be dropping rows instead.
# inplace=True: This argument tells the drop() method to modify the DataFrame directly. If you set inplace=False, the drop() method would return a new DataFrame with the columns removed, leaving the original DataFrame unchanged.
# So, in essence, this code removes the latitude and longitude columns from the "covid_conf" DataFrame, modifying the DataFrame directly.
```

```python
covid_conf.head()
```

```python
agg_df = covid_conf.groupby("Country/Region").sum()
```

```python
agg_df.shape
```

```python
#Data viz  resepected to countries



# The respective row contains non numeric objects, hence converted to number values then 
china_data = agg_df.loc["China"]
china_data.plot()


afgan_data = agg_df.loc["Afghanistan"]
afgan_data.plot()
india_data = agg_df.loc["India"]
india_data.plot()
spain_data = agg_df.loc["Spain"]
spain_data.plot()


plt.legend()
```

```python
agg_df.isnull()

```

```python
#calulating the first derivative
china_data.diff().plot()
india_data.diff().plot()
afgan_data.diff().plot()
spain_data.diff().plot()

plt.legend()
```

```python
#Add a col to show max infect rates among all the countries

countries = list(agg_df.index)
max_inf_rates = []
for c in countries : 
    max_inf_rates.append(agg_df.loc[c].diff().max())
    
agg_df["max_infected_rate"] = max_inf_rates
    
```

```python
agg_df.head()
```

```python
# new df

co_data = pd.DataFrame(agg_df["max_infected_rate"] )
co_data
```

```python
hap_rep = pd.read_csv("/kaggle/input/covid-19-analysis-data/worldwide_happiness_report.csv")
```

```python
hap_rep
```

```python
hap_rep.drop(["Overall rank", "Score", "Generosity","Perceptions of corruption" ],axis=1, inplace = True )
```

```python
hap_rep
```

```python
hap_rep.set_index('Country or region', inplace = True)
```

```python
hap_rep.head()
```

```python
joined_df = co_data.join(hap_rep,how="inner")
```

```python
sb.heatmap(joined_df.corr(),annot=True, cmap="RdYlGn", fmt=".2f", linewidths=0.5)
```

```python

x = joined_df["GDP per capita"]
y = joined_df["max_infected_rate"]

plt.figure(figsize=(10, 6))
sb.scatterplot(x=x, y=np.log(y), color="blue", marker="o")

plt.xlabel("GDP per Capita")
plt.ylabel("Log of Max Infected Rate")
plt.title("Scatterplot of GDP per Capita vs Log of Max Infected Rate")
plt.grid(True) 
plt.tight_layout()
plt.show()
```

```python

x = joined_df["GDP per capita"]
y = joined_df["max_infected_rate"]

# Plotting
plt.figure(figsize=(10, 6))
sb.regplot(x=x, y=np.log(y), color="blue", marker="o")

# Adding labels and title
plt.xlabel("GDP per Capita")
plt.ylabel("Log of Max Infected Rate")
plt.title("Regressed Scatterplot of GDP per Capita vs Log of Max Infected Rate")
plt.grid(True)  # Optional for better visualization
plt.tight_layout()
plt.show()
```
